### Description of the feature used to identify the phishing website.

#### How to identify phishing website from the URL ? 

Attackers changes the subdomain and file path(If it contains in the URL) or making the typo mistake to look like the legitimate website. So to identify the phishing website mainly we need to breakdown the URL to look what each parts of the URL contains. 

What is the out of scop for the given dataset? 
Attackers can use the direct IP address instead of using the domain name. 
Attackers can make typo mistake or very small domain name which looks like the legitimate one.

```python
def impute_mode(df, columns):
    for column in columns:
        print((df[column].values == -1).sum())
        mode0 = round(df[df[column].values != -1][df["phishing"] == 0][column].mode()[0])
        print(mode0)
        mode1 = round(df[df[column].values != -1][df["phishing"] == 1][column].mode()[0])
        print(mode1)
        mask = (df[column] == -1) & (df['phishing'] == 0)
        df[column][mask] = mode0
        mask = (df[column] == -1) & (df['phishing'] == 1)
        df[column][mask] = mode1
        print((df[column].values == -1).sum())
    return df
```

### Replace noise from the feature (Utility functions)

```python
def impute_mean(df, columns):
    for column in columns:
        print((df[column].values == -1).sum())
        mean0 = round(df[df[column].values != -1][df["phishing"] == 0][column].mean())
        print(mean0)
        mean1 = round(df[df[column].values != -1][df["phishing"] == 1][column].mean())
        print(mean1)
        mask = (df[column] == -1) & (df['phishing'] == 0)
        df[column][mask] = mean0
        mask = (df[column] == -1) & (df['phishing'] == 1)
        df[column][mask] = mean1
        print((df[column].values == -1).sum())
    return df
```

```python
def impute_median(df, columns):
    for column in columns:
        print((df[column].values == -1).sum())
        median0 = round(df[df[column].values != -1][df["phishing"] == 0][column].median())
        print(median0)
        median1 = round(df[df[column].values != -1][df["phishing"] == 1][column].median())
        print(median1)
        mask = (df[column] == -1) & (df['phishing'] == 0)
        df[column][mask] = median0
        mask = (df[column] == -1) & (df['phishing'] == 1)
        df[column][mask] = median1
        print((df[column].values == -1).sum())
    return df
```

```python
def replace_mode(cat_column, phishing, value_0, value_1):
    if cat_column == -1 and phishing==0:
        cat_column=value_0
    elif cat_column == -1 and phishing==1:
        cat_column=value_1
    else:
        cat_column=cat_column
    return cat_column
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import GridSearchCV

model = LogisticRegression(max_iter=1000)
solvers = ['newton-cg', 'lbfgs', 'liblinear']
penalty = [None, 'l2']
c_values = [100, 10, 1.0, 0.1, 0.01]

import time
start_time=time.time()

grid = dict(solver=solvers,penalty=penalty,C=c_values)
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
grid_search_lr = GridSearchCV(estimator=model, param_grid=grid, n_jobs=-1, cv=cv, scoring='accuracy',error_score=0)
grid_result_lr = grid_search_lr.fit(X_train, y_train)
print("Best: %f using %s" % (grid_result_lr.best_score_, grid_result_lr.best_params_))
means = grid_result_lr.cv_results_['mean_test_score']
stds = grid_result_lr.cv_results_['std_test_score']
params = grid_result_lr.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
end_time=time.time()   
print('Time taken is ', (end_time-start_time), ' secs'  )
y_pred_lr = grid_search_lr.predict(X_test)
print("MSE : ", mean_squared_error(y_test, y_pred_lr))
print(classification_report(y_test, y_pred_lr))
print("Accuracy: ", accuracy_score(y_test, y_pred_lr))
cm = confusion_matrix(y_test, y_pred_lr)
ax = plt.axes()
sns.heatmap(cm, annot=True, annot_kws={"size": 10}, fmt='d',cmap="Blues", ax = ax )
ax.set_title('Confusion Matrix for logistic regression.')
plt.show()


from sklearn import datasets, metrics, model_selection
metrics.plot_roc_curve(grid_search_lr, X_test, y_test)
plt.show()
```
