### Description of the feature used to identify the phishing website.

#### How to identify phishing website from the URL ? 

Attackers changes the subdomain and file path(If it contains in the URL) or making the typo mistake to look like the legitimate website. So to identify the phishing website mainly we need to breakdown the URL to look what each parts of the URL contains. 

What is the out of scop for the given dataset? 
Attackers can use the direct IP address instead of using the domain name. 
Attackers can make typo mistake or very small domain name which looks like the legitimate one.

```python
def impute_mode(df, columns):
    for column in columns:
        print((df[column].values == -1).sum())
        mode0 = round(df[df[column].values != -1][df["phishing"] == 0][column].mode()[0])
        print(mode0)
        mode1 = round(df[df[column].values != -1][df["phishing"] == 1][column].mode()[0])
        print(mode1)
        mask = (df[column] == -1) & (df['phishing'] == 0)
        df[column][mask] = mode0
        mask = (df[column] == -1) & (df['phishing'] == 1)
        df[column][mask] = mode1
        print((df[column].values == -1).sum())
    return df
```

### Logistic Regression
